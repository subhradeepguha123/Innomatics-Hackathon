```python
df=df.drop_duplicates()
df
```

```python
df=df.drop(['Unnamed: 0', 'Id', 'groupId', 'matchId'],axis=1)
df
```

```python
df.info()
```

```python
obj_df=df.select_dtypes(include=['object']).copy()
obj_df
```

```python
num_df=df.select_dtypes(exclude=['object']).copy()
num_df
```

```python
obj_df.value_counts()
```

```python
sns.countplot(y="matchType",data=obj_df)
```

```python
for i in num_df.columns:
    sns.distplot(num_df[i])
    plt.show()
```

```python
list1= obj_df.columns
list1
```

```python
from sklearn.preprocessing import LabelEncoder
def Encoder():
    for i in list1:
        Encoder=LabelEncoder()
        obj_df[i]=Encoder.fit_transform(obj_df[i])
Encoder()
```

```python
d = dict.fromkeys(obj_df.select_dtypes(np.int32).columns, np.int64)
obj_df = obj_df.astype(d)
print (obj_df.dtypes)
```

```python
df_raw=pd.concat([num_df,obj_df],axis=1,join='inner')
df_raw
```

```python
ax = sns.clustermap(df_raw.corr(), annot=True, linewidths=.6, fmt= '.2f', figsize=(20, 15))
plt.show()
```

```python
for i in num_df.columns:
    sns.boxplot(num_df[i])
    plt.show()
```

```python
plt.figure(figsize=(15,8))
ax1 = sns.boxplot(x="kills",y="damageDealt", data = num_df)
ax1.set_title("Damage Dealt vs. Number of Kills")
plt.show()
```

```python
plt.figure(figsize=(15,8))
ax3 = sns.boxplot(x="DBNOs",y="kills", data = num_df)
ax3.set_title("NUmber of DBNOs vs. Number of Kills")
plt.show()
```
